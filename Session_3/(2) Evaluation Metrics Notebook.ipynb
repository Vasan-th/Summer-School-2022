# Common Performance Metrics in Machine Learning
Performance metrics are a part of every machine learning pipeline. They tell you if you’re making progress, and put a number on it.
Choosing the right metric is crucial while evaluating machine learning (ML) models. All machine learning models, whether it’s regression or classification, needs a metric to judge performance.
Usually in a Kaggle competition more often than not, the final leaderboard is based on a performance metric.

Performance metrics are divided into **classification metrics** and **Regression Metrics**.

Regression metrics include **Mean Squared Error** and **Mean Absolute Error**.

In this session we will focus on classification Metrics.



# Classification Metrics

Classification is one of the most widely used problems in machine learning with various industrial applications, from face recognition, Youtube video categorization, content moderation, medical diagnosis, to text classification, hate speech detection on Twitter.

To judge a classification task we have several metrics which are listed below
*   Classification Accuracy
*   Precision 
*   Recall
*   F1 Score




# Confusion Matrix
One of the key concept in classification performance is confusion matrix (AKA error matrix), which is a tabular visualization of the model predictions versus the ground-truth labels. Each row of confusion matrix represents the instances in Ground Truth class and each column represents the instances in Predicted class.


image.png

In order to understand the confusion matrix, we need to set some value for the null hypothesis as an assumption. For example, from our Breast Cancer data, let’s assume our Null Hypothesis H⁰ be **“The individual has cancer”**.

Capture_wiki.PNG

Each cell in the confusion matrix represents an evaluation factor. Let’s understand these factors one by one:


*   **True Positive(TP)** 
**  signifies how many positive class samples your model predicted correctly.
**  The individual has cancer and doctors have predicted that he has cancer.

*   **True Negative(TN)** 
**  signifies how many negative class samples your model predicted correctly. 
**  The individual doesnt have cancer and the doctors have predicted that he doesnt have cancer.

*   **False Positive(FP)** 
**  signifies how many negative class samples your model predicted incorrectly. This factor represents **Type-I** error in statistical nomenclature. 
**  The individual doesnt have cancer but the doctors have predicted that he has cancer

*   **False Negative(FN)** 
**  signifies how many positive class samples your model predicted incorrectly. This factor represents **Type-II** error in statistical nomenclature. 
**  The individual has cancer but the doctor has deemed him cancer free.

```python
#implementing confusion matrix
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
y_true_ex = [2, 0, 2, 2, 0, 1]
y_pred_ex = [0, 0, 2, 2, 0, 2]

#creating confusion matrix
conf_mat_ex=confusion_matrix(y_true_ex, y_pred_ex)
fig, ax = plot_confusion_matrix(conf_mat=conf_mat_ex, figsize=(6, 6), cmap=plt.cm.Greens)
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

# Classification Accuracy
Classification accuracy is defined as the number of correct predictions divided by the total number of predictions, multiplied by 100.

Accuracy=$\frac{True Positive + True Negative}{Total Number of Predictions}$


```python
from sklearn.metrics import accuracy_score
print(f'Accuracy Score of svm model is {accuracy_score(y_test,y_pred)}')
print(f'Accuracy Score of naive bayes model is {accuracy_score(Y_test,Y_pred)}')
print(f'Accuracy Score of knn model is {accuracy_score(Y_test_KNN,Y_pred_KNN)}')
```

# Precision
Precision is the ratio of true positives and total positives predicted

Precision=$\frac{True Positive}{True Positive + False Positive}$

0<P<1
 
For the previous example on cancer

Precision=$\frac{TP}{TP + FP}$=$\frac{Cancerous\  patients\  predicted\  as\  cancerous}{Cancerous\  patients\  predicted\  as\  cancerous\ + \ Non\ Cancerous\ patients\ predicted \ as\ cancerous}$

The precision metric focuses on Type-I errors(FP). A Type-I error occurs when we reject a true null Hypothesis.

A higher Precision value(close to 1) indicates very low amount of Type 1 error




```python
from sklearn.metrics import precision_score
print(f'precision Score of svm model is {precision_score(y_test,y_pred)}')
print(f'precision Score of naive bayes model is {precision_score(Y_test,Y_pred, average = None)}')
print(f'precision Score of knn model is {precision_score(Y_test_KNN,Y_pred_KNN, average = None)}')
```

# Recall
A Recall is essentially the ratio of true positives to all the positives in ground truth. It is also known as True Positivity Rate.

Recall=$\frac{TP}{TP + FN}$=$\frac{Cancerous\  patients\  predicted\  as\  cancerous}{Cancerous\  patients\  predicted\  as\  cancerous\ + \ Cancerous\ patients\ predicted \ as\ non\  cancerous}$

0<R<1

A low recall score (<0.5) means your classifier has a high number of false negatives which can be an outcome of imbalanced class or untuned model hyperparameters. In an imbalanced class problem, you have to prepare your data beforehand with over/under-sampling or focal loss in order to curb FP/FN.

The recall metric focuses on type-II errors(FN). So, in this case, type-II error is incorrectly labeling cancerous patients as non cancerous. 

```python
from sklearn.metrics import recall_score
print(f'recall Score of svm model is {recall_score(y_test,y_pred)}')
print(f'recall Score of naive bayes model is {recall_score(Y_test,Y_pred, average = None)}')
print(f'recall Score of knn model is {recall_score(Y_test_KNN,Y_pred_KNN, average = None)}')
```

# ROC curve (Receiver Operating Characteristic curve)
An ROC curve (receiver operating characteristic curve) is a graph showing the performance of a classification model at all classification **thresholds**.

So what is a Threshold?

A Threshold is a specified limit in the probabilities computed for each data point(features) for each class.

For example check the dataset below, setting different thresholds for classifying positive class for data points will inadvertently change the False Positivity Rate and Recall of the model. And one of these thresholds will probably give a better result than the others, depending on whether we are aiming to lower the number of False Negatives or False Positives.

image.png

This curve plots two parameters:

* True Positive Rate(Recall)
* False Positive Rate
